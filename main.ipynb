# Math 441 Convex Optimization

## Problem statement

For an electical vehicle driving on the highway, the driver need to charge the car at the charging station during the travel, we want to figure out how the driver choose which station to chagring such that the vehicle can drive through the highway in a minimum time, and the loss of battery lifetime can also be minimized.

## Variables and Parameters
|   Symbol   | Description                                                          |         Type         |
|:----------:| -------------------------------------------------------------------- |:--------------------:|
| $i$ | index of charging station | index |
| $x_i$ | equal to 1 if the driver choose to charge at station $i$, otherwise not|  decision variable  |
| $e_i$ | the amount of energy charged at station $i$ | decision variable |
| $c_i$ | equal to 1 if the driver choose fast charging at station $i$, otherwise normal charging| decision variable |
| $v$ | the speed of the vehicle driving in highway | parameter |
| $D_i$ | the distance need to drive from start point or last charging station to station $i$ | parameter |
| $B_{\text{max}}$ | the battery capacity of the vehicle | parameter |
| $l$ | the damage (capacity loss) to the battery | parameter |
