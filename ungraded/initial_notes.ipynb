## Idea 2: Airline Delay-Insurance Design

design delay-insurance terms for a set of airlines by choosing:
- the premium charged for each airline policy
- the compensation paid if a delay occurs

goal is to balance:
- customer attractiveness
- expected payout cost
- risk from high compensation exposure
- fairness across airlines

### Sets and Indices

Let

$$
A = \{1,2,\dots,n\}
$$

be the set of airlines.

For each airline $i \in A$:

- $T_i$ = ticket price for airline $i$
- $q_i$ = delay probability for airline $i$
- $k_i > 0$ = premium sensitivity parameter
- $m_i > 0$ = compensation sensitivity parameter
- $\bar{x}_i$ = upper bound on premium fraction
- $\bar{y}_i$ = upper bound on compensation fraction


### Decision Variables

$$
x_i = \text{premium charged as a fraction of ticket price for airline } i
$$

$$
y_i = \text{compensation paid as a fraction of ticket price for airline } i
$$

$$
P_i = x_i T_i
$$

$$
C_i = y_i T_i
$$

where:

- $P_i$ is the premium charged
- $C_i$ is the compensation paid if a delay occurs


### Interpretation of the Objective

We use the term

$$
-k_i P_i + m_i C_i
$$

as a simplified attractiveness score for the policy of airline $i$:

- higher premium reduces attractiveness
- higher compensation increases attractiveness

So for airline $i$:

$$
-k_i P_i + m_i C_i = -k_i x_i T_i + m_i y_i T_i
$$

This is not exact profit, but a convex proxy for how attractive the policy is to customers.


### Model 1: Baseline Linear Program

####  Objective Function

Maximize total policy attractiveness:

$$
\max_{x,y} \quad \sum_{i \in A} \left(-k_i x_i T_i + m_i y_i T_i\right)
$$

#### Constraints

Expected compensation budget:

$$
\sum_{i \in A} q_i y_i T_i \le B
$$

Premium bounds:

$$
0 \le x_i \le \bar{x}_i, \qquad \forall i \in A
$$

Compensation bounds:

$$
0 \le y_i \le \bar{y}_i, \qquad \forall i \in A
$$

Consistency constraint:

$$
x_i \le y_i, \qquad \forall i \in A
$$

#### Full Baseline Model

$$
\begin{aligned}
\max_{x,y} \quad & \sum_{i \in A} \left(-k_i x_i T_i + m_i y_i T_i\right) \\
\text{subject to} \quad
& \sum_{i \in A} q_i y_i T_i \le B \\
& 0 \le x_i \le \bar{x}_i, \qquad \forall i \in A \\
& 0 \le y_i \le \bar{y}_i, \qquad \forall i \in A \\
& x_i \le y_i, \qquad \forall i \in A
\end{aligned}
$$

- the objective is linear in $x_i$ and $y_i$
- all constraints are linear

Therefore it is a convex optimization problem.


## Model 2: Risk-Aware Convex Quadratic Model

### Motivation

The baseline model controls only the **expected payout**:

$$
\sum_{i \in A} q_i y_i T_i
$$

But this does not penalize concentration in risky policies.

To make the model more realistic, we add a convex quadratic risk penalty:

$$
\lambda \sum_{i \in A} q_i (y_i T_i)^2
$$

where:

- $\lambda > 0$ is the risk-aversion parameter
- larger $\lambda$ means the optimizer penalizes generous compensation more strongly for risky airlines

### Objective Function

We now maximize:

$$
\max_{x,y} \quad
\sum_{i \in A} \left(-k_i x_i T_i + m_i y_i T_i\right)
-
\lambda \sum_{i \in A} q_i (y_i T_i)^2
$$

### Constraints

Keep the same constraints as the baseline model:

$$
\sum_{i \in A} q_i y_i T_i \le B
$$

$$
0 \le x_i \le \bar{x}_i, \qquad \forall i \in A
$$

$$
0 \le y_i \le \bar{y}_i, \qquad \forall i \in A
$$

$$
x_i \le y_i, \qquad \forall i \in A
$$

### Full Risk-Aware Model

$$
\begin{aligned}
\max_{x,y} \quad
& \sum_{i \in A} \left(-k_i x_i T_i + m_i y_i T_i\right)
- \lambda \sum_{i \in A} q_i (y_i T_i)^2 \\
\text{subject to} \quad
& \sum_{i \in A} q_i y_i T_i \le B \\
& 0 \le x_i \le \bar{x}_i, \qquad \forall i \in A \\
& 0 \le y_i \le \bar{y}_i, \qquad \forall i \in A \\
& x_i \le y_i, \qquad \forall i \in A
\end{aligned}
$$

The term

$$
\sum_{i \in A} q_i (y_i T_i)^2
$$

is convex because:

- each $(y_i T_i)^2$ is a convex quadratic function
- $q_i \ge 0$
- weighted sums of convex functions are convex

Since the objective is being maximized, it is equivalent to maximizing a concave function or minimizing a convex one.  
The feasible region is still defined by linear constraints.

So this is a convex quadratic optimization problem.


## Model 3: Risk-Aware Model with Fairness / Minimum Attractiveness

### Motivation

Without extra constraints, the model may produce policies that are attractive for some airlines and very poor for others.

To avoid that, require each airline policy to achieve at least a minimum attractiveness level.

Let $L_i$ be the minimum attractiveness required for airline $i$.

Then impose:

$$
-k_i x_i T_i + m_i y_i T_i \ge L_i, \qquad \forall i \in A
$$

This ensures that every airline's policy remains at least somewhat competitive.

### Full Model

$$
\begin{aligned}
\max_{x,y} \quad
& \sum_{i \in A} \left(-k_i x_i T_i + m_i y_i T_i\right)
- \lambda \sum_{i \in A} q_i (y_i T_i)^2 \\
\text{subject to} \quad
& \sum_{i \in A} q_i y_i T_i \le B \\
& 0 \le x_i \le \bar{x}_i, \qquad \forall i \in A \\
& 0 \le y_i \le \bar{y}_i, \qquad \forall i \in A \\
& x_i \le y_i, \qquad \forall i \in A \\
& -k_i x_i T_i + m_i y_i T_i \ge L_i, \qquad \forall i \in A
\end{aligned}
$$

The new constraint

$$
-k_i x_i T_i + m_i y_i T_i \ge L_i
$$

is linear.

So the objective remains concave for maximization, and all constraints remain linear.

Thus Model 3 is still convex.


### Optional Extension: Diversification Constraint

To avoid extremely uneven compensation levels across airlines, we can impose a norm constraint:

$$
\|y\|_2 \le \delta
$$

where:

- $y = (y_1,\dots,y_n)^T$
- $\delta > 0$ controls how concentrated compensation levels can be

This is convex because norm constraints are convex.

If included, the model becomes:

$$
\begin{aligned}
\max_{x,y} \quad
& \sum_{i \in A} \left(-k_i x_i T_i + m_i y_i T_i\right)
- \lambda \sum_{i \in A} q_i (y_i T_i)^2 \\
\text{subject to} \quad
& \sum_{i \in A} q_i y_i T_i \le B \\
& 0 \le x_i \le \bar{x}_i, \qquad \forall i \in A \\
& 0 \le y_i \le \bar{y}_i, \qquad \forall i \in A \\
& x_i \le y_i, \qquad \forall i \in A \\
& -k_i x_i T_i + m_i y_i T_i \ge L_i, \qquad \forall i \in A \\
& \|y\|_2 \le \delta
\end{aligned}
$$
